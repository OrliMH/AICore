# DECISION TREE

A  decision tree is a decision support tool that uses a tree-like model of decisions and  
their possible consequences, including chance event outcomes, resource costs, and utility.  
It is one way to display an algorithm that only contains conditional control statements.

# DECISION TREE LEARNING

Decision tree learning is one of the predictive modelling aproaches used in  
statistics, data mining and machine learnin. It uses a decision tree(as a  
predictive model) to go from observations about an item(represented in the  
branches) to conclusions about the item's target value(represented in leaves).  
Tree models where the target variable can take a discrete set of values are  
called classification trees; Decision trees where the target variables can  
take continuous values(typically real numbers) are called regression trees.

# Metrics

Algorithms for constructing decision trees usually work top-down, by choosing  
a variable at each step that best splits the set of items. Metrics generally  
measure the homogeneity of the target variables within the subsets.  

### Gini impurity(基尼不纯度)  
#### 越低越好  
Used by the CART(classification and regression tree) algorithm for classification  
trees, Gini impurity is a measure of how often a randomly chosen element from the  
set would be incorrectly labeled if it was randomly labeled according to the distribution  
of labels in the subset. 
To compute Gini impurity for a set of items with J classes, suppose $i \in \lbrace1,2,...,J\rbrace$,  
and let $p_i$ be the fraction of items labeled with class $i$ in the set.  


$$
\begin{align}
I_{G(p)}\ =\ \sum_{i=1}^{J}p_i\sum_{k\neq i}p_k\ =\ \sum_{i=1}^{J}p_i(1-p_i) \\
       =\ \sum_{i=1}^{J}(p_i-p_i^2)\ =\ \sum_{i=1}^{J}p_i-\sum_{i=1}^{J}p_i^2\ =\ 1-\sum_{i=1}^{J}p_i^2
\end{align}
$$

### Information gain(信息增益）
越低越好  
Entropy is defined as below
$$H(T)\ =\ I_E(p_1,p_2,...,p_J)\ =\ -\sum_{i=1}^{J}p_i\log_2{p_i},\ p_i\ is\ the\ problability\ of\ class\ i\ in\ a\ node(target\ variable)$$

### Advantages  
1. clear to explain   
2. could select salient features(显著特征）  
It can be used to select salient features to decrease the dimensions of models with redundant features. Press Enter twice to quit list.


### Disadvantages  
1. so sentitive  异常点放在树的上方，容易过拟合  
The algorithm can be easily effected by outliers or exceptional items which may cause overfitting.  
2. limited fitting ability  维度高需要深数 分布均匀找不到强特征  
It may needs a deep tree structure to fit when the dimension is high. When the data is uniformly distrubuted,    
it is hard to find salient features.  

### Loss of decision tree  
$$Loss\ =\ \frac{m_{left}}{m}*G_{left}\ +\ \frac{m_{right}}{m}*G_{right},\ m\ is\ the\ scale\ of\ a\ tree$$

# SELECT SALIENT FEATURES OF BOSTEN HOUSING PRICE WITH DECISION TREE

```python
from sklearn.datasets import load_boston
boston = load_boston()
X = boston.data
y = boston.target
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_clf = DecisionTreeRegressor()
tree_clf.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_clf, out_file = 'boston.dot', feature_names = boston.feature_names, rounded=True,
               filled=True)
for line in open('boston.dot'):
    print(line)
```

```python
tree_clf.feature_importances_
```

```python
{n:w for n,w in zip(boston.feature_names, tree_clf.feature_importances_)}
```

```python
sorted({n:w for n,w in zip(boston.feature_names, tree_clf.feature_importances_)}.items(), key= lambda x:x[1], reverse=True)
```

# UNDERSTANDING GINI IMPURITY AND INFORMATION GAIN

```python
import numpy as np
from collections import Counter
```

```python
def pr(e, es):
    return Counter(es)[e]/len(es)
```

```python
def gini(elements):
    return 1-np.sum(pr(e, elements)**2 for e in set(elements))
    
```

```python
features_1 = ['R','R','Y','Y']
features_2 = ['R','R','R','Y']
features_3 = ['R','R','R','R']
```

```python
gini(features_1)
```

```python
gini(features_2)
```

```python
gini(features_3)
```

```python
def entropy(elements):
    return -np.sum(pr(e, elements)*np.log2(pr(e, elements)) for e in set(elements))
```

```python
entropy(features_1)
```

```python
entropy(features_2)
```

```python
entropy(features_3)
```
